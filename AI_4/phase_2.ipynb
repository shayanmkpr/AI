```python
import pandas as pd
import numpy as np
import scipy.stats as sci 
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import tree
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import cross_val_score
```

```python
df = pd.read_csv(r"D:\#10\AI\CA4\house_data.csv")
df.info()
df['sqft_living'] = df['sqft_living'].fillna(df['sqft_living'].mode()[0])
df['floors'] = df['floors'].fillna(df['floors'].mode()[0])
df['sqft_basement'] = df['sqft_basement'].fillna(df['sqft_basement'].mode()[0])
df['yr_built'] = df['yr_built'].fillna(df['yr_built'].mode()[0])
```

```python
for i in df.columns:
    label_encoder = preprocessing.LabelEncoder()
    if df[i].dtype == 'object':
        df[i]= label_encoder.fit_transform(df[i])
        df[i].unique()
df
```

```python
median = df['price'].median()
df['price_level'] = ['1' if x >= median else '-1' for x in df['price']]
```

```python


X = df.drop('price_level', axis=1)
y = df['price_level']
max_depth_range = [3, 5, 7, 10, 15, 20, 25]
cv_scores = []
for max_depth in max_depth_range:
    dt = DecisionTreeRegressor(max_depth=max_depth)
    scores = cross_val_score(dt, X, y, cv=5, scoring='neg_mean_squared_error')
    cv_scores.append(scores.mean())
for i in range(len(max_depth_range)):
    print("Max depth: {}, CV score: {}".format(max_depth_range[i], cv_scores[i]))
max_score_index = cv_scores.index(max(cv_scores))
best_max_depth = max_depth_range[max_score_index]
print("Best max depth:", best_max_depth)
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop('price_level', axis=1), df['price_level'], test_size=0.2)
max_depth_range = [2, 3, 4, 5, 6]
min_samples_split_range = [2,3, 5, 10, 15, 20]
criterion_range = ['gini', 'entropy']
best_accuracy = 0
best_max_depth = None
best_min_samples_split = None
best_criterion = None
for max_depth in max_depth_range:
    for min_samples_split in min_samples_split_range:
        for criterion in criterion_range:
            clf = tree.DecisionTreeClassifier(max_depth=max_depth, min_samples_split=min_samples_split, criterion=criterion)
            clf.fit(X_train, y_train)
            y_pred_test = clf.predict(X_test)
            test_accuracy = accuracy_score(y_test, y_pred_test)
            if test_accuracy > best_accuracy:
                best_accuracy = test_accuracy
                best_max_depth = max_depth
                best_min_samples_split = min_samples_split
                best_criterion = criterion
print('Best max depth:', best_max_depth)
print('Best min samples split:', best_min_samples_split)
print('Best criterion:', best_criterion)
print('Best accuracy:', best_accuracy)
cm = confusion_matrix(y_test, y_pred_test)
print('Confusion Matrix:\n', cm)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
X_train, X_test, y_train, y_test = train_test_split(df.drop('price_level', axis=1), df['price_level'], test_size=0.2)
k_range = list(range(1, 10))
best_accuracy = 0
best_k = None
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred_test = knn.predict(X_test)
    test_accuracy = accuracy_score(y_test, y_pred_test)
    if test_accuracy > best_accuracy:
        best_accuracy = test_accuracy
        best_k = k
cm = confusion_matrix(y_test, y_pred_test)
print('Confusion Matrix:\n', cm)
print('Best k:', best_k)
print('Best accuracy:', best_accuracy)
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop('price_level', axis=1), df['price_level'], test_size=0.2)
param_grid = {'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000]}
lr = LogisticRegression()
lr_cv = GridSearchCV(lr, param_grid, cv=5)
lr_cv.fit(X_train, y_train)
y_pred_test = lr_cv.predict(X_test)
cm = confusion_matrix(y_test, y_pred_test)
print('Confusion Matrix:\n', cm)
test_accuracy = accuracy_score(y_test, y_pred_test)
print('Test accuracy:', test_accuracy)

print('Optimal hyperparameters:', lr_cv.best_params_)
```
